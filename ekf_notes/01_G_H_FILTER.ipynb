```python
# NOW LETS TRY A RANDOMLY CHOSE NUMBER TO SCALE OUR ESTIMATE SAY 4/10. 
# OUR ESTIMATE WIL BE 4/10'S THE MEASRUEMTN AND THE REST WILL BE FROM THE PREDICTION. 
# IN OTHER WORDS WE ARE EXPRESSSING A BELIEF HERE, A BELIEF THAT THE PREDTICTIN IS SOMEWHT MORE LIKELY TO BE CORRECT THAN THE MEASUREMENT. 
# THIS BELIEF CAN BE EXPRESSES AS --> NEW_ESTIMATE = PREDICTION + (4/10)*(MEASUREMENT-PREDICTION)

# THE DIFFERENCE BETWEEN THE MEASUREMENT AND PREDICITON IS CALLED THE RESIDUAL , WHICH IS DEPICTED BY THE BLACK VERTICAL LINE IN THE PLOT ABOVE. 
# THIS WILL BECOME AN IMPORTANT VALUE TO USE LATER ON, AS IT IS AN EXACT COMPUTATION OF THE DIFFERENCE BETWEEN MEASUREMENTS AND THE FILTER'S OUTPUT. 
# SMALLER RESIDUALS IMPLY BETTER PERFORMANCE 
```

```python
weights = [158.0, 164.2, 160.3, 159.9, 162.1, 164.6, 
           169.6, 167.4, 166.4, 171.0, 171.2, 172.6]

from __future__ import division, print_function  # this was needed to have the scale factor perform its magic
time_step = 1 # day
scale_factor = 4/10
print("scale factor {}".format(scale_factor))

def predict_using_gain_guess(weight, gain_rate, do_print=True): 
    # store the filtered results
    estimates = []
    predictions = []
    estimates.append(weight)

    # most filter literature uses 'z' for measurements
    for z in weights: 
        # predict new position
        prediction = weight + gain_rate * time_step

        # update filter 
        weight = prediction + scale_factor * (z - prediction)

        # save for plotting
        estimates.append(weight)
        predictions.append(prediction)
        if do_print:
            print('previous: {:.2f}, prediction: {:.2f} estimate {:.2f}'.format(
                  estimates[-2], prediction, weight))
            
            
    # plot results
    n = len(weights)
    xs = list(range(n+1))
    
    # plot_filter

    
    
    
    lines = plt.plot(xs,estimates, color='C0', label='Filter', marker='o')
    
    # plot measurements
    plt.autoscale(tight=False)
    plt.plot(xs[1:], weights, color='b', lw=2, ls='--', label='Scale')

    
    # plot_track
    plt.plot([0, n], [160, 160+n], color='k', lw=2, ls=':', label='Actual Weight')
    
    # plot_track
    plt.plot(xs[1:], predictions, color='r', lw=2, ls=':', label='Predictions', marker='v')
    
    
    # other stuff
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
    plt.xlabel('day')
    plt.ylabel('weight (lbs)')
    
    plt.xlim([0, n])
    
    plt.show()
    
initial_guess = 160.

predict_using_gain_guess(weight=initial_guess, gain_rate=1)
    
```

```python
#  THE TOP ISNT SO USEFUL BECAUSE IT REQUIRES US TO CORRECTLY GUESS A RATE OF CHAGNE . 
# EVEN IF OUR INITIAL GUESS WAS CORRECT, THTE FILTER WILL FAIL AS SOON AS THE RATE OF CHANGE CHAGNES. 


# below is an example of the g-h algorithm 
# g and h refer to the two scaling factors that are usess
# g is the caling we used for the measurement (weight in our example)

# h is the caling for the change in measuremetn over time 
```

```python
weight = 160  # initial guess
gain_rate = 1.0 # initial guess

time_step = 1
weight_scale = 4/10
gain_scale = 1/3
estimates = [weight]
predictions = []


for z in weights:
    # prediction step
    weight = weight + gain_rate*time_step
    gain_rate = gain_rate
    predictions.append(weight)
    
    # update step    
    residual = z - weight
    
    gain_rate = gain_rate + gain_scale   * (residual/time_step)
    weight    = weight    + weight_scale * residual
  
    estimates.append(weight)
    
# plot results
n = len(weights)
plt.xlim([1, n])
days = np.linspace(1, n, n)

# PLOT_FILTER
lines = plt.plot(range(len(estimates)), estimates, color='C0')

# PLOT_MEASUREMETNS
plt.autoscale(tight=False)
plt.plot(days, weights, color='b', lw=2, ls='--', label='Scale')

# PLOT_TRACK
plt.plot([0, n], [160, 160+n], color='k', lw=2, ls=':', label='Actual Weight')



# PLOT_TRACK

plt.plot(range(1, n+1), predictions, color='r', lw=2, ls=':', label='Predictions')

plt.xlabel('day')
plt.ylabel('weight (lbs)')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))


    
plt.xlim([1, n])
    
plt.show()
```

```python
# The g-h Filter

# This algorithm is known as the g-h filter.
# g and h refer to the two scaling factors that we used in our example.
# g is the scaling we used for the measurement (weight in our example), and h is the scaling for the change in measurement over time (lbs/day in our example).


# This filter is the basis for a huge number of filters, including the Kalman filter. 
# In other words, the Kalman filter is a form of the g-h filter, which I will prove later in the book. 
# So is the Least Squares filter, which you may have heard of, and so is the Benedict-Bordner filter, which you probably have not.
# Each filter has a different way of assigning values to g and h, but otherwise the algorithms are identical. 
# For example, the \alpha-\beta filter assigns a constant to g and h, constrained to a certain range of values. 
# Other filters such as the Kalman will vary g and h dynamically at each time step.
```
